```python
def count_indicators(comments, feature_words, feature_words2, val, val2):
    num_of_feature_words = len(feature_words)
    num_of_feature_words2 = len(feature_words2)

    
    indicators = []
    for comment in comments:
        counter = 0
        indicator_sum = 0.
        comment_split = comment.split() 
        for word in comment_split:
            if word in feature_words:
                indicator_sum += (num_of_feature_words * val)
                counter += num_of_feature_words
            elif word in feature_words2:
                indicator_sum += (num_of_feature_words2 * val2)
                counter += num_of_feature_words2
            else:
                words = feature_words + feature_words2
                for ft_word in words:
                    counter += 1
                    dist = models["twitter_25d"].calculate_distance(ft_word, word)
                    indicator_sum += dist
        indicator_avrg = indicator_sum / counter
        indicators.append(indicator_avrg)
#         print(f"Positive index for sent\n{comment}:\n {indicator_avrg}")
#         print(f"{indicator_avrg}")
    return indicators
```

```python
def grade(indicator, thresholds):
    if indicator > thresholds[7]:
        return 7
    elif indicator > thresholds[6]:
        return 6
    elif indicator > thresholds[5]:
        return 5
    elif indicator > thresholds[4]:
        return 4
    elif indicator > thresholds[3]:
        return 3
    elif indicator > thresholds[2]:
        return 2
    else:
        return 1
```

```python
def count_indicator(comment, feature_words, feature_words2, val, val2, thresholds):
    num_of_feature_words = len(feature_words)
    num_of_feature_words2 = len(feature_words2)

    counter = 0
    indicator_sum = 0.
    comment_split = comment.split() 
    for word in comment_split:
        if word in feature_words:
            indicator_sum += (num_of_feature_words * val)
            counter += num_of_feature_words
#             print(val)
        elif word in feature_words2:
            indicator_sum += (num_of_feature_words2 * val2)
            counter += num_of_feature_words2
#             print(val)
        else:
            for ft_word in feature_words:
                counter += 1
                dist = models["twitter_25d"].calculate_distance(ft_word, word)
#                 print(dist)
                indicator_sum += dist
#         print(indicator_sum)
#     print(indicator_sum / counter)
    return grade(indicator_sum / counter, thresholds)
```
